# Web Scraping Code File

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
features=[]
rating=[]
phone_name=[]
price=[]

for i in range(1,40):
    start_time=time.time()
    URL="https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&sort=popularity&page={}".format(i)
    page=requests.get(URL)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    for x in soup.find_all('div',attrs={'class':'_2kHMtA'}):
        pname=x.find('div',attrs={'class':'_4rR01T'})
        rat=x.find('div',attrs={'class':'_3LWZlK'})
        cost=x.find('div',attrs={'class':'_30jeq3 _1_WHN1'}) 
        feat=x.find('ul',attrs={'class':'_1xgFaf'})
        if pname is None:
            phone_name.append(np.NaN)
        else:
            phone_name.append(pname.text)
            
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if rat is None:
            rating.append(np.NaN)
        else:
            rating.append(rat.text)
            
        if feat is None:
            features.append(np.NaN)
        else:
            features.append(feat.text)
    
        print('Page{} completed in {} seconds'.format(i,time.time()-start_time))
```

```python
mobile_df=pd.DataFrame({'Model_Name':phone_name,'Price':price,'Features':features,'Rating':rating})
mobile_df
```

```python
regex1= re.compile(r'\d+\s\w\w\sRAM')
regex2= re.compile(r'\d+\s\w\w\sROM')
regex3= re.compile(r'\d+\smAh')
mobile_df['RAM']=mobile_df['Features'].apply(lambda x:re.compile(regex1).findall(x))
mobile_df['ROM']=mobile_df['Features'].apply(lambda x:re.compile(regex2).findall(x))
mobile_df['Battery']=mobile_df['Features'].apply(lambda x:re.compile(regex3).findall(x))
regex6=re.compile(r'Expandable Upto \d+\s[A-Z]+')
mobile_df['Expandable_Memory']=mobile_df['Features'].apply(lambda x:re.compile(regex6).findall(x))
mobile_df.head()
```

```python
regex4=re.compile(r'\d+MP\sFront\sCamera')
mobile_df['Front_Cam']=mobile_df['Features'].apply(lambda x:re.compile(regex4).findall(x))
regex5=re.compile(r'Battery[\w\s]+Processor')
mobile_df['Processor']=mobile_df['Features'].apply(lambda x:re.compile(regex5).findall(x))
regex7=re.compile(r'\d+.\d+\scm[\S\s]+Display')
mobile_df['Display']=mobile_df['Features'].apply(lambda x:re.compile(regex7).findall(x))
mobile_df[0:1]
```

```python
mobile_df['RAM']=mobile_df['RAM'].apply(lambda x:''.join(x))
mobile_df['ROM']=mobile_df['ROM'].apply(lambda x:''.join(x))
mobile_df['Battery']=mobile_df['Battery'].apply(lambda x:''.join(x))
mobile_df['Expandable_Memory']=mobile_df['Expandable_Memory'].apply(lambda x:''.join(x))
mobile_df['Front_Cam']=mobile_df['Front_Cam'].apply(lambda x:''.join(x))
mobile_df['Processor']=mobile_df['Processor'].apply(lambda x:''.join(x))
mobile_df['Display']=mobile_df['Display'].apply(lambda x:''.join(x))
mobile_df[0:2]
```
